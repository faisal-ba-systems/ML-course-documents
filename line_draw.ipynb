```python
# Re-import required libraries since execution state was reset
import numpy as np
import matplotlib.pyplot as plt

# Original 1D data points
X = np.array([-4, -3, -2, -1, 0, 1, 2, 3, 4])
y = np.array([-1, -1, -1, -1, 1, 1, 1, 1, 1])  # Two classes (-1 and 1)

# First Plot: 1D Points
plt.figure(figsize=(8, 4))
plt.scatter(X, np.zeros_like(X), c=y, cmap='coolwarm', edgecolors='k', s=100)
plt.axhline(y=0, color='black', linestyle='-')
plt.xlabel("X-axis (Original 1D Data)")
plt.title("1D Data Points")
plt.xticks(X)
plt.yticks([])
plt.grid()
plt.show()

# Transform the points by squaring them
X_transformed = X**2

# Second Plot: 2D Transformation
plt.figure(figsize=(8, 6))
plt.scatter(X, X_transformed, c=y, cmap='coolwarm', edgecolors='k', s=100)
plt.xlabel("X-axis (Original Data)")
plt.ylabel("Squared Feature X")
plt.title("Data Transformation (Squaring the Points)")

# Decision boundary (A separating line)
x_line = np.linspace(-4, 4, 100)
plt.plot(x_line, x_line**2 - 5, 'k--', label="Decision Boundary")

plt.legend()
plt.grid()
plt.show()
```

```python
# Define new labels for color coding
labels = np.array([0, 0, 1, 1, 1, 1, 1, 0, 0])

# Define colors based on labels (0 = Blue, 1 = Red)
colors = np.array(['blue' if lbl == 0 else 'red' for lbl in labels])

# First Plot: 1D Points with Labels
plt.figure(figsize=(8, 4))
for i, (x, lbl) in enumerate(zip(X, labels)):
    plt.scatter(x, 0, color=colors[i], edgecolors='k', s=100 ,label=labels)
    # plt.text(x, 0.1, str(lbl), fontsize=12, ha='center', color=colors[i])  # Label above the point

plt.axhline(y=0, color='black', linestyle='-')
plt.xlabel("X-axis (Original 1D Data)")
plt.title("1D Data Points with Labels")
# plt.xticks(X)
# plt.yticks([])
plt.grid()
plt.show()
```

```python
# First Plot: 1D Points with Labels and a Legend
plt.figure(figsize=(8, 4))

# Plot points with unique labels
for lbl, color in zip([0, 1], ['blue', 'red']):
    plt.scatter([], [], color=color, label=f"Class {lbl}")  # Legend entry

# Plot points
for i, (x, lbl) in enumerate(zip(X, labels)):
    plt.scatter(x, 0, color=colors[i], edgecolors='k', s=100)
    # plt.text(x, 0.1, str(lbl), fontsize=12, ha='center', color=colors[i])  # Label above the point

plt.axhline(y=0, color='black', linestyle='-')
plt.xlabel("X-axis (Original 1D Data)")
plt.title("1D Data Points with Labels and Legend")

# Add legend
plt.legend()
plt.grid()
plt.show()
```

```python

# Transform the points by squaring them
X_transformed = X**2

# Second Plot: 2D Transformation with Labels
plt.figure(figsize=(8, 6))
for i, (x, x_sq, lbl) in enumerate(zip(X, X_transformed, labels)):
    plt.scatter(x, x_sq, color=colors[i], edgecolors='k', s=100)
    plt.text(x, x_sq + 0.5, str(lbl), fontsize=12, ha='center', color=colors[i])  # Label above the point

plt.xlabel("X-axis (Original Data)")
plt.ylabel("Squared Feature X")
plt.title("Data Transformation (Squaring the Points) with Labels")

# Decision boundary (A separating line)
x_line = np.linspace(-4, 4, 100)
# plt.plot(x_line, x_line**2 - 5, 'k--', label="Decision Boundary")
plt.plot(x_line, x_line**2 - 5, 'k--', label="Decision Boundary")

plt.legend()
plt.grid()
plt.show()
```

```python
# Second Plot: 2D Transformation with Labels and a Straight Decision Line
plt.figure(figsize=(8, 6))
for i, (x, x_sq, lbl) in enumerate(zip(X, X_transformed, labels)):
    plt.scatter(x, x_sq, color=colors[i], edgecolors='k', s=100)
    plt.text(x, x_sq + 0.5, str(lbl), fontsize=12, ha='center', color=colors[i])  # Label above the point

plt.xlabel("X-axis (Original Data)")
plt.ylabel("Squared Feature X")
# plt.title("Data Transformation (Squaring the Points) with a Straight Decision Line")
plt.title("Squaring the Points")
# Straight Decision Boundary (Horizontal Line)
plt.axhline(y=6, color='green', linestyle='--', label="Decision Boundary")

plt.legend()
plt.grid()
plt.show()
```
